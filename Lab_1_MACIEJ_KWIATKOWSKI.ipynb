NumPy udostępnia strukturę `ndarray`, która:

* jest szybka dzięki ciągłemu układowi w pamięci
* pozwala na wektoryzację obliczeń

## ✅ Podstawowe atrybuty tablicy

```python
a = np.array([0, 1, 2, 3])

print(type(a))
print(a.dtype)
print(a.shape)
print(a.ndim)
```

### 📘 Teoria

* `type(a)` → typ obiektu (ndarray)
* `dtype` → typ danych (np. int64)
* `shape` → wymiary tablicy
* `ndim` → liczba wymiarów

---

## ✅ Operacje element‑po‑elemencie

```python
a = np.array([1, 2, 3, 4])
b = np.array([2, 3, 4, 5])

print(a + b)
print(a * b)
print(a ** b)
```

### 📘 Teoria

NumPy domyślnie wykonuje operacje **element‑wise**:

[ c_i = f(a_i, b_i) ]

---

## ✅ Tablice wielowymiarowe

```python
a = np.array([[0, 1, 2, 3],
              [10, 11, 12, 13]])

print(a.shape)
print(a.size)
print(a.ndim)
print(a[1])
```

### 📘 Teoria

* `shape` → (wiersze, kolumny)
* `size` → liczba wszystkich elementów
* `a[1]` → drugi wiersz

---

## ✅ Indeksowanie i slicing

```python
a = np.array([10, 11, 12, 13, 14])

print(a[1:3])
print(a[1:-2])
print(a[:3])
print(a[-2:])
print(a[::2])
```

### 📘 Teoria

Slicing: `[start:stop:step]`

---

## ⚠️ Widoki vs kopie danych

```python
a = np.array([0, 1, 2, 3, 4])

b = a[2:4]
b[0] = 10

print(a)
print(b)
```

### 📘 Teoria

Slicing tworzy **widok**, nie kopię → zmiany wpływają na oryginał.

---

## ✅ Maski logiczne

```python
a = np.arange(0, 80, 10)
mask = np.array([0, 1, 1, 0, 0, 1, 0, 0], dtype=bool)

b = a[mask]
print(a)
print(mask)
print(b)
```

### 📘 Teoria

Maski pozwalają filtrować dane.

---

## ✅ Operacje wzdłuż osi

```python
a = np.array([[1, 2, 3],
              [4, 5, 6]])

print(a.sum())
print(a.sum(axis=0))
print(a.sum(axis=1))
print(np.max(a, axis=1))
print(np.argmax(a, axis=1))
print(np.mean(a, axis=0))
```
|
### 📘 Teoria

`axis=0` → kolumny
`axis=1` → wiersze

---

## ✅ np.where()

```python

mask = a % 2 == 0
b = np.where(mask)

```

### 📘 Teoria

`where` zwraca indeksy spełniające warunek.

## ✅ Iloczyn skalarny


# 🧠 Python list vs NumPy vs PyTorch

## 📌 Python list

✔ Struktura ogólnego przeznaczenia

✔ Może przechowywać różne typy danych

✔ Operacje wykonywane wolniej

✔ Brak operacji wektorowych

```python
lst = [1, 2, 3]
```

❗ Operacje są wykonywane element po elemencie w pętli (nawet jeśli tego nie widzimy).

---

## 📌 NumPy array

✔ Dane jednego typu (`dtype`)

✔ Bardzo szybkie obliczenia

✔ Operacje wektorowe (element‑wise)

✔ Idealne do obliczeń numerycznych

```python
arr = np.array([1, 2, 3])
```

✔ NumPy używa zoptymalizowanego kodu w C → duża wydajność.

---

## 📌 PyTorch Tensor

✔ Podobny do NumPy array

✔ Obsługa GPU

✔ Automatyczne różniczkowanie (gradienty)

✔ Fundament sieci neuronowych

```python
tensor = torch.tensor([1.0, 2.0, 3.0])
```

---

## ⚡ Kluczowe różnice

| Cecha              | Python list | NumPy         | PyTorch       |
| ------------------ | ----------- | ------------- | ------------- |
| Typ danych         | dowolne     | jednorodne    | jednorodne    |
| Szybkość           | wolna       | bardzo szybka | bardzo szybka |
| GPU                | ❌           | ❌             | ✔             |
| Operacje wektorowe | ❌           | ✔             | ✔             |
| Gradienty          | ❌           | ❌             | ✔             |

# 🧪 Mini‑eksperyment wydajnościowy

👉 Zadanie: Uruchom kod i porównaj czasy.

````python
import time
import numpy as np
import torch

N = 10_000_000

print("Generowanie danych...")
lst = list(range(N))
arr = np.array(lst)
tensor_cpu = torch.tensor(lst)

# -------------------------
# Python list
# -------------------------
start = time.time()
_ = [x * 2 for x in lst]
print("Python list:", time.time() - start)

# -------------------------
# NumPy
# -------------------------
start = time.time()
_ = arr * 2
print("NumPy:", time.time() - start)

# -------------------------
# PyTorch CPU
# -------------------------
start = time.time()
_ = tensor_cpu * 2
print("PyTorch CPU:", time.time() - start)

# -------------------------
# PyTorch GPU
# -------------------------
if torch.cuda.is_available():
    print("\nTest GPU...")

    tensor_gpu = tensor_cpu.to('cuda')

    torch.cuda.synchronize()
    start = time.time()

    _ = tensor_gpu * 2

    torch.cuda.synchronize()
    print("PyTorch GPU:", time.time() - start)
else:
    print("\nGPU niedostępne")
````

### 🧠 Zadanie 13:

👉 Jaki będzie dtype?

👉 Dlaczego?

👉 Dlaczego to ma znaczenie w sieciach neuronowych?

````python
import torch

x = torch.tensor([1, 2, 3])
y = torch.tensor([1.0, 2.0, 3.0])

z = x + y

print(z.dtype)
````

````python
dtype będzie torch.float32
Tensor x ma typ int64, a tensor y ma typ float32.
Podczas wykonywania operacji PyTorch stosuje automatyczną promocję typów. Oznacza to, że gdy łączymy typ całkowity z typem zmiennoprzecinkowym, wynik zostaje zamieniony na float, aby nie utracić części dziesiętnej.
W sieciach neuronowych większość operacji wykorzystuje liczby zmiennoprzecinkowe (najczęściej float32)
````
